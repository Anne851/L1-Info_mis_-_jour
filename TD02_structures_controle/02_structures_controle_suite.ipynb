**Exercice 1 :** Supposons que vous souhaitiez développer un programme permettant à un élève de première année de s'entraîner à la soustraction. Le programme génère de manière aléatoire deux entiers d'un seul chiffre, number1 et number2, avec number1 >= number2, et pose à l'élève une question telle que "Quel est 9 - 2 ?" Après que l'élève ait saisi la réponse, le programme affiche un message indiquant si elle est correcte. Ecrire un programme qui génère cinq questions et, après qu'un élève y ait répondu, rapporte le nombre de réponses correctes. Le programme affiche également le temps passé sur le test, comme le montre l'exécution d'exemple. Pour mesurer le temps, il faut importer la bibliothèque *Time* (exemple d'utilisation: temps_de_depart = time.time()).

```python
import random
import time

def generer_question():
    number1 = random.randint(1, 9)
    number2 = random.randint(0, number1)
    return number1, number2

def poser_question(number1, number2):
    print("Quel est",number1, "-", number2, "?")
    reponse = int(input("Votre réponse : "))
    return reponse == (number1 - number2)


def main():
    nombre_questions = 5
    bonnes_reponses = 0

    print("Bienvenue dans l'exercice de soustraction !")
    input("Appuyez sur Entrée pour commencer")

    temps_debut = time.time()

    for i in range(nombre_questions):
        print(f"\nQuestion {i+1}:")
        n1, n2 = generer_question()
        if poser_question(n1, n2):
            print("Correct !")
            bonnes_reponses += 1
        else:
            print("Incorrect. La bonne réponse était {n1 - n2}.")

    temps_fin = time.time()
    temps_total =(temps_fin - temps_debut)

    print("\nVous avez terminé le test !")
    print("Nombre de bonnes réponses :", bonnes_reponses, "sur", nombre_questions)
    print("Temps total :",temps_total, "secondes")
main()
```

**Exercice 2 :**  L'exemple précédent exécute la boucle cinq fois. Si vous souhaitez que l'utilisateur décide s'il souhaite prendre une autre question, vous pouvez proposer une confirmation à l'utilisateur (en tapant 'Y' pour continuer).

```python
import random
import time
continuer =True


def generer_question():
    number1 = random.randint(1, 9)
    number2 = random.randint(0, number1)
    return number1, number2

def poser_question(number1, number2):
    print("Quel est",number1, "-", number2, "?")
    reponse = int(input("Votre réponse : "))
    return reponse == (number1 - number2)

def demander_continuation():
    continuer = input("Taper Y pour continuer")
    if continuer == 'Y':
        continuer =True
    else:
        continuer = False
    return continuer


def main():
    nombre_questions = 0
    bonnes_reponses = 0
    continuer = True

    print("Bienvenue dans l'exercice de soustraction !")
    input("Appuyez sur Entrée pour commencer.")

    temps_debut = time.time()

    
    while continuer == True:
        nombre_questions += 1
        print("\nQuestion",nombre_questions)
        n1, n2 = generer_question()
        if poser_question(n1, n2):
            print("Correct !")
            bonnes_reponses += 1
        else:
            print("Incorrect. La bonne réponse était", n1 - n2)
        
        continuer = demander_continuation()
    temps_fin = time.time()
    temps_total = (temps_fin - temps_debut, )

    print("\nVous avez terminé le test !")
    print("Nombre de bonnes réponses :", bonnes_reponses, "sur", nombre_questions)
    print("Temps total :",temps_total,"secondes")
main()

```

**Exercice 4 :**  Écrivez un programme qui joue au populaire jeu ciseaux-pierre-papier. (Un ciseau peut couper du papier, une pierre peut écraser un ciseau, et du papier peut envelopper une pierre.) Le programme génère aléatoirement un nombre 0, 1 ou 2, représentant respectivement ciseaux, pierre et papier. Ensuite, le programme demande à l'utilisateur d'entrer un nombre 0, 1 ou 2, puis affiche un message indiquant si l'utilisateur gagne, perd ou obtient un match nul par rapport à l'ordinateur. Le programme  permet à l'utilisateur de jouer en continu jusqu'à ce que l'utilisateur ou l'ordinateur gagne plus de deux fois.

```python
import random

def generer_choix_ordinateur():
    return random.randint(0, 2)

def obtenir_choix_utilisateur():
    while True:
        choix = input("Entrez votre choix (0 pour ciseaux, 1 pour pierre, 2 pour papier) : ")
        if choix in ['0', '1', '2']:
            return int(choix)
        else: print("Choix invalide. Veuillez entrer 0, 1 ou 2.")

def determiner_gagnant(choix_utilisateur, choix_ordinateur):
    if choix_utilisateur == choix_ordinateur:
        return "égalité"
    elif (choix_utilisateur == 0 and choix_ordinateur == 2) or \
         (choix_utilisateur == 1 and choix_ordinateur == 0) or \
         (choix_utilisateur == 2 and choix_ordinateur == 1):
        return "utilisateur"
    else:
        return "ordinateur"

def afficher_choix(choix):
    options = ["ciseaux", "pierre", "papier"]
    return options[choix]

def main():
    score_utilisateur = 0
    score_ordinateur = 0
    
    print("Bienvenue au jeu Ciseaux-Pierre-Papier!")
    print("Le premier à gagner plus de deux fois remporte la partie.")

    while score_utilisateur <= 2 and score_ordinateur <= 2:
        choix_utilisateur = obtenir_choix_utilisateur()
        choix_ordinateur = generer_choix_ordinateur()

        print("Vous avez choisi :", afficher_choix(choix_utilisateur))
        print("L'ordinateur a choisi :", afficher_choix(choix_ordinateur))

        resultat = determiner_gagnant(choix_utilisateur, choix_ordinateur)

        if resultat == "utilisateur":
            print("Vous gagnez cette manche!")
            score_utilisateur += 1
        elif resultat == "ordinateur":
            print("L'ordinateur gagne cette manche!")
            score_ordinateur += 1
        else:
            print("C'est une égalité!")

        print("Score actuel - Vous :", score_utilisateur, "Ordinateur :" , score_ordinateur)

    if score_utilisateur > score_ordinateur:
        print("Félicitations! Vous avez gagné la partie!")
    else:
        print("L'ordinateur a gagné la partie.")
main()
```

**Exercice 5 :** Écrivez un programme pour vérifier la validité des mots de passe saisis par les utilisateurs.

    -Au moins 1 lettre entre [a-z] et 1 lettre entre [A-Z].
    -Au moins 1 chiffre entre [0-9].
    -Au moins 1 caractère parmi [$#@].
    -Longueur minimale de 6 caractères.
    -Longueur maximale de 16 caractères.
    
La fonction *re.search()* du module *re* en Python est utilisée pour rechercher un motif (expression régulière) dans une chaîne de caractères. Voici la syntaxe simplifiée de re.search() : re.search(pattern, string)

    -pattern : C'est le motif (expression régulière) que vous souhaitez rechercher dans la chaîne de caractères.
    -string : C'est la chaîne de caractères dans laquelle vous souhaitez effectuer la recherche.
    
La fonction re.search() renvoie un objet de correspondance (match object) si le motif est trouvé dans la chaîne de caractères. Si le motif n'est pas trouvé, elle renvoie None.    

```python
import re

def verifier_mot_de_passe(mot_de_passe):
    if len(mot_de_passe) < 6 or len(mot_de_passe) > 16:
        return False

    if not re.search(r"[a-z]", mot_de_passe):
        return False

    if not re.search(r"[A-Z]", mot_de_passe):
        return False

    if not re.search(r"[0-9]", mot_de_passe):
        return False

    if not re.search(r"[$#@]", mot_de_passe):
        return False

    return True

def main():
    print("Vérificateur de mot de passe")
    print("Le mot de passe doit respecter les critères suivants :")
    print("- Au moins 1 lettre entre [a-z] et 1 lettre entre [A-Z]")
    print("- Au moins 1 chiffre entre [0-9]")
    print("- Au moins 1 caractère parmi [$#@]")
    print("- Longueur minimale de 6 caractères")
    print("- Longueur maximale de 16 caractères")
    
    while True:
        mot_de_passe = input("\nEntrez un mot de passe (ou 'q' pour quitter) : ")
        
        if mot_de_passe == 'q':
            print("Au revoir!")
            break
        
        if verifier_mot_de_passe(mot_de_passe):
            print("Mot de passe valide!")
        else:
            print("Mot de passe invalide. Veuillez réessayer.")
main()
```
